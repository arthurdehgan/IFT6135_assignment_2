```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
def epoch_time(path):
    """return the list of time spent of every epoch of training phase of a model given the path of the folder of the model"""
    with open(path) as f:
        content = f.readlines()
    return [float(a.split("spent in epoch: ")[1]) for a in content]

def cumulative_epoch_time(epoch_time):
    return [np.sum(epoch_time[:i+1]) for i in range(len(epoch_time))]
```

## 4.1
#### Experiment 1 - RNN
<br>
Hyperparameters : <br>
<ul>
    <li>Model : RNN</li>
    <li>Optimizer : ADAM</li>
    <li>Initial learning rate :0.0001</li>
    <li>Batch size : 20</li>
    <li>Sequense length : 35</li>
    <li>Hidden size : 1500</li>
    <li>Number of layers : 2</li>
    <li>Dropout keep probability : 0.35</li>
</ul>
Command : <br>
<pre>--model=RNN --optimizer=ADAM --initial_lr=0.0001 --batch_size=20 --seq_len=35 --hidden_size=1500 --num_layers=2 --dp_keep_prob=0.35 --save_best</pre>

```python
path = "./results/PB4_1/RNN_ADAM_model=RNN_optimizer=ADAM_initial_lr=0.0001_batch_size=20_seq_len=35_hidden_size=1500_num_layers=2_dp_keep_prob=0.35_save_best_3/"
tmp = np.load(path+"learning_curves.npy")[()]
```

```python
plt.plot([i for i in range(len(tmp["train_ppls"]))],np.log10(tmp["train_ppls"]), label="Train")
plt.plot([i for i in range(len(tmp["train_ppls"]))],np.log10(tmp["val_ppls"]), label="Test")
plt.legend()
plt.title("Perplexity per epoch")
plt.ylabel("$log_{10}$ of perplexity")
plt.xlabel("Epoch")
```

```python
et = epoch_time(path+"log.txt")
cet = cumulative_epoch_time(et)
plt.plot(cet,np.log10(tmp["train_ppls"]), label="Train")
plt.plot(cet,np.log10(tmp["val_ppls"]), label="Test")
plt.legend()
plt.title("Perplexity per time")
plt.ylabel("$log_{10}$ of perplexity")
plt.xlabel("Time (s)")
```
